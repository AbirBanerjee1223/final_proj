```python

import pandas as pd
from Bio.SeqUtils.ProtParam import ProteinAnalysis
from collections import Counter
import math

# STEP 1: Load your dataset
try:
    df = pd.read_excel("ang.xlsx", engine="openpyxl")
except:
    df = pd.read_csv("ang.csv")

print("‚úÖ File loaded successfully! Total rows:", len(df))

# STEP 2: Define helper functions
def aliphatic_index(seq):
    aa = Counter(seq)
    A, V, I, L = aa.get('A',0), aa.get('V',0), aa.get('I',0), aa.get('L',0)
    total = sum(aa.values())
    if total == 0:
        return 0
    return 100 * (A/total + 2.9*(V/total) + 3.9*((I+L)/total))

def sequence_entropy(seq):
    """Shannon entropy: diversity of amino acid composition"""
    aa_counts = Counter(seq)
    total = sum(aa_counts.values())
    if total == 0:
        return 0
    probs = [count/total for count in aa_counts.values()]
    return -sum(p * math.log2(p) for p in probs if p > 0)

def amino_acid_composition(seq):
    """Returns % composition of each amino acid"""
    aa_counts = Counter(seq)
    total = sum(aa_counts.values())
    if total == 0:
        return {aa: 0 for aa in valid_aas}
    return {aa: (aa_counts.get(aa,0)/total)*100 for aa in valid_aas}

# STEP 3: Prepare DataFrame for new columns
features = [
    "Molecular_Weight", "Isoelectric_Point", "Aromaticity",
    "Instability_Index", "Aliphatic_Index", "Net_Charge_pH7",
    "Hydrophobicity_GRAVY", "Positive_Residues", "Negative_Residues",
    "Shannon_Entropy"
]
for col in features:
    df[col] = None

valid_aas = set("ARNDCQEGHILKMFPSTWYV")

# Add amino acid % columns
for aa in valid_aas:
    df[f"AAC_{aa}"] = None

# STEP 4: Process each sequence
for i, seq in enumerate(df["Sequence"]):
    try:
        seq = str(seq).upper().replace(" ", "")
        seq = seq.replace("B", "N").replace("Z", "Q").replace("X", "")
        seq = "".join([aa for aa in seq if aa in valid_aas])

        if len(seq) == 0:
            print(f"‚ö† Row {i}: empty or invalid sequence, skipped.")
            continue

        analysis = ProteinAnalysis(seq)

        # Basic physicochemical properties
        df.loc[i, "Molecular_Weight"] = analysis.molecular_weight()
        df.loc[i, "Isoelectric_Point"] = analysis.isoelectric_point()
        df.loc[i, "Aromaticity"] = analysis.aromaticity()
        df.loc[i, "Instability_Index"] = analysis.instability_index()
        df.loc[i, "Aliphatic_Index"] = aliphatic_index(seq)
        df.loc[i, "Net_Charge_pH7"] = analysis.charge_at_pH(7.0)
        df.loc[i, "Hydrophobicity_GRAVY"] = analysis.gravy()

        aa = Counter(seq)
        df.loc[i, "Positive_Residues"] = aa.get("K",0) + aa.get("R",0) + aa.get("H",0)
        df.loc[i, "Negative_Residues"] = aa.get("D",0) + aa.get("E",0)

        # New features
        df.loc[i, "Shannon_Entropy"] = sequence_entropy(seq)
        comp = amino_acid_composition(seq)
        for aa in valid_aas:
            df.loc[i, f"AAC_{aa}"] = comp[aa]

    except Exception as e:
        print(f"‚ùå Error at row {i}: {e}")

# STEP 5: Save to new Excel file
output_file = "protein_features_extended.xlsx"
df.to_excel(output_file, index=False)

print("\n‚úÖ Feature calculation complete!")
print(f"üíæ Results saved to: {output_file}")
print("Rows with valid sequences:", df['Molecular_Weight'].notna().sum())
print("Rows skipped:", len(df) - df['Molecular_Weight'].notna().sum())
```

```python
import pandas as pd

# --- 1. Load the Essential Genes List ---
# Create sets for both CEGs and CSEGs for fast lookup

ceg_genes = set()
cseg_genes = set()

try:
    # We need to find the header row first, as done previously
    with open('CSEGs_CEGs.txt', 'r') as f:
        all_lines = f.readlines()

    header_line = "gene\tessentiality\tncbi_id\tensembl\n"
    header_index = -1
    for i, line in enumerate(all_lines):
        if line == header_line:
            header_index = i
            break
    
    if header_index == -1:
        print("Error: Could not find the header line in 'CSEGs_CEGs.txt'.")
    else:
        # Load the CSV, skipping rows before the header
        essential_df = pd.read_csv(
            'CSEGs_CEGs.txt',
            sep='\t',
            skiprows=header_index,
            header=0
        )
        
        # Populate the sets
        ceg_genes = set(essential_df[essential_df['essentiality'] == 'CEGs']['gene'])
        cseg_genes = set(essential_df[essential_df['essentiality'] == 'CSEGs']['gene'])
        
        print(f"‚úÖ Loaded {len(ceg_genes)} CEGs and {len(cseg_genes)} CSEGs into lookup sets.")

except FileNotFoundError:
    print("Error: The file 'CSEGs_CEGs.txt' was not found.")
    exit()
except Exception as e:
    print(f"An error occurred while loading essential genes: {e}")
    exit()


# --- 2. Load Your Main 20,420 Protein Dataset ---
# 
# -----------------------------------------------------------------
# !!!  IMPORTANT  !!!
# REPLACE "random_fet.csv" WITH THE FILENAME OF YOUR *MAIN* DATASET
# This file MUST contain the 'Gene Names' column.
# For example: "protein_features_extended.csv"
# -----------------------------------------------------------------
main_dataset_filename = "final_training_dataset.csv" 

try:
    df_main = pd.read_csv(main_dataset_filename)
    print(f"‚úÖ Loaded main dataset '{main_dataset_filename}' with {len(df_main)} rows.")
except FileNotFoundError:
    print(f"Error: The main data file '{main_dataset_filename}' was not found.")
    print("Please update the 'main_dataset_filename' variable in the code.")
    exit()
except Exception as e:
    print(f"An error occurred while loading the main dataset: {e}")
    exit()

# --- 3. Define the Labeling Function ---
def classify_essentiality(gene_names_str):
    """
    Checks if any gene in the cell is a CEG or CSEG.
    Prioritizes CEGs.
    """
    if not isinstance(gene_names_str, str):
        return 'Non-Essential' # Handle empty/NaN rows
    
    gene_list = gene_names_str.split() 
    
    # Check for CEGs first
    for gene in gene_list:
        if gene in ceg_genes:
            return 'CEG' # Core-Essential
    
    # If no CEG, check for CSEGs
    for gene in gene_list:
        if gene in cseg_genes:
            return 'CSEG' # Cancer-Specific Essential
    
    return 'Non-Essential' # Not in either list

# --- 4. Create the 'Essentiality_Type' Column ---
if 'Gene Names' not in df_main.columns:
    print("Error: The column 'Gene Names' was not found in your main dataset.")
    print("This column is required to perform the check.")
else:
    print("\nLabeling proteins based on gene names...")
    df_main['Essentiality_Type'] = df_main['Gene Names'].apply(classify_essentiality)
    
    # --- 5. Check the Results ---
    print("\n‚úÖ Labeling complete!")
    print("--------------------------------------------------")
    print("Distribution of proteins in your dataset:")
    print("--------------------------------------------------")
    # This will print the counts
    print(df_main['Essentiality_Type'].value_counts())
```
